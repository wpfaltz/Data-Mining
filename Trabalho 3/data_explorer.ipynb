```python
import os
import pandas as pd

# Definir o diretório base
base_dir = r"C:\Users\wpett\Downloads\Trabalho 3 - Data Mining"

# Inicializar dicionários para armazenar os tamanhos mínimos e máximos globais
columns = ['ordem', 'latitude', 'longitude', 'datahora', 'velocidade', 'linha', 'datahoraenvio', 'datahoraservidor']
global_min_lengths = {col: float('inf') for col in columns}
global_max_lengths = {col: 0 for col in columns}

# Percorrer o diretório e subdiretórios
for root, dirs, files in os.walk(base_dir):
    for file in files:
        if file.startswith('2024') and file.endswith('json'):
            file_path = os.path.join(root, file)
            try:
                # Ler o arquivo JSON
                df = pd.read_json(file_path)
                # Garantir que todos os dados sejam strings
                df = df.astype(str)
                for col in columns:
                    # Calcular os tamanhos mínimos e máximos da coluna atual
                    min_length = df[col].str.len().min()
                    max_length = df[col].str.len().max()
                    # Atualizar os valores globais
                    if min_length < global_min_lengths[col]:
                        global_min_lengths[col] = min_length
                    if max_length > global_max_lengths[col]:
                        global_max_lengths[col] = max_length
                print(f"Lido com sucesso: {file_path}")
                # Liberar memória do DataFrame
                del df
            except Exception as e:
                print(f"Erro ao ler {file_path}: {e}")

# Forçar a coleta de lixo para liberar memória não utilizada
import gc
gc.collect()

# Exibir os resultados globais
print("Tamanhos mínimos globais:")
for col, length in global_min_lengths.items():
    print(f"{col}: {length}")

print("\nTamanhos máximos globais:")
for col, length in global_max_lengths.items():
    print(f"{col}: {length}")
```
